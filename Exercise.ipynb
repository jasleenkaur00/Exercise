```python
import pandas as pd
from math import pow, sqrt

#Columns Names
m_cols = ['userid', 'name', 'gender', 'age', 'interest']

#Read Data from file into datadrame
users = pd.read_csv('dating.txt', sep='::', names=m_cols, encoding='latin-1', engine='python')

#print(users)

#Split Interest columns
users.interest = users.interest.str.split('|')

#Unique values of Interests
interest_columns = list(set([j for i in users['interest'].tolist() for j in i]))
#print(interest_columns)

#Convert Interest from delimitted string to Interest columns and mark 0 or 1
for j in interest_columns:
    users[j] = 0
for i in range(users.shape[0]):
    for j in interest_columns:
        if(j in users['interest'].iloc[i]):
            users.loc[i,j] = 1
            
#print(users)

#Create copy and insert Score column
user_score = users.copy()
user_score.insert(column="score",value=0.0000,loc=17)
# drop redundant comma separated interest column
users.drop('interest',axis=1,inplace=True)

#print(user_score)

#Function to create ranking of users that match the input user id
def CalculateRanking(input_user_id) :

    for i, row in user_score.iterrows():
        score = 0
        #set score zero for self so that self does not come as match
        if row['userid'] == input_user_id :
            user_score.at[i,'score'] = 0
        else:
            #Difference of age
            ageFactor = row['age'] - user_score.loc[input_user_id, 'age']

            #opposite gender
            if row['gender'] == user_score.loc[input_user_id, 'gender']:
                genderFactor=50 #given weightage can be adjusted to completely exclude same gender match
            else :
                genderFactor=0
                
            
            #Interest 
            interestFactor=0
            for j in interest_columns:
                if row[j] == 1 and  user_score.loc[input_user_id, j] == 1 :
                    interestFactor +=  1

            #score=pow(ageFactor,2)+pow(genderDiff,2)+pow((1/(1+interestFactor)),2)
            score=pow(ageFactor,2)+pow(genderFactor,2)-pow(interestFactor,2)


            ranking = (1/(1+sqrt(score)))
            #print( user_score.loc[input_user_id, 'name'], row['name'],i," age factor " , ageFactor, " gender Factor ", genderFactor, " interest factor ",  interestFactor," score ",score ," ranking ", ranking)


            user_score.at[i,'score'] = ranking

#USE THIS FUNCTION TO CALCULATE RELEVANT RANKING BASED ON USERID INPUT AS PARAMETER
#INPUT PARAMETER IS THE USERID IN THE TXT FILE
CalculateRanking(0)
#print(user_score)

user_score.sort_values(by=['score'],ascending=False)
#print(user_score)
    
```
